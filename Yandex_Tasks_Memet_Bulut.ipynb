```python

import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt

df = pd.read_csv("movie_metadata.csv")

df.head(50)
```

Answer of Question # 1.2. The duration column contains data on the film length. How many missing values are there in this column?

```python
duration_missing_values = df['duration'].isnull().sum()
print(duration_missing_values)
```

Answer of Question #  1.3. Replace the missing values in the `duration` column with the median value for this column.

```python
median_duration = df['duration'].median()
print(median_duration)
```

```python
# median_duration_with_median = df['duration'].replace(np.NaN, median_duration)
# print(median_duration_with_median)
```

```python
df["duration"] = df['duration'].replace(np.NaN, median_duration)
```

Answer of # 1.4. What is the average film length? Give the answer as a floating-point figure rounded to two decimal places.

```python
film_length_average = df['duration'].mean().round(decimals=2)
print(film_length_average)
```

Answer of # 1.5. How many films between 90 minutes and two hours long were released in 2008?

```python
 df.loc[(df['duration'] >= 90) & (df['duration'] <= 120) & (df['title_year'] == 2008.0) ]
```

Answer of # 1.6. The budget column contains the film's budget. What is the median budget for all the films listed? Give the answer as an integer.

```python
budget = df['budget'].replace('[\$,]', '', regex=True).astype(float)

budget.median().astype(int)
```
